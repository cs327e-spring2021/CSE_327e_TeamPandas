# Load in Census Data (ACS)

### Before we load in we have to adjust some data fields!

For example, for ACS data all rows do not display what state they belong too. Instead they have dividers only. We'll fix using pandas and then save as csv before saving and loading in.

### Housing and demographic data 

```python
# Read in data and veiw
import pandas as pd
df=pd.read_csv('ACS_Demographic_and_hosuing_data.csv')
# Reformat rows
# Remove all percentage rows
df.drop(df[df['Label'] =='Percent'].index, inplace = True)

# create new coloumn
df['state']=''

# Add state column
for index, row in df.iterrows():
    if df.loc[index][0] != 'Estimate':
        state_name=df.loc[index][0]
    else:
        df.loc[index]['state']=state_name
        
# remove spacers
df.drop(df[df['Label'] !='Estimate'].index, inplace = True)
df.drop(labels=['Label'],axis=1, inplace = True)
df.to_csv('ACS_demo_house_mod.csv', index=False)
df.head()
```

We can see that the label column isn't corrrect- has an empty row + "estimate" and "percent" don't show state. We can also remove percent since it's just a repeat of estimate but converted.

### Social Charcteric data

```python
# Read in data and veiw
import pandas as pd
df=pd.read_csv('ACS_selected_social_char.csv')
# Reformat rows
# Remove all percentage rows
df.drop(df[df['Label'] =='Percent'].index, inplace = True)

# create new coloumn
df['state']=''

# Add state column
for index, row in df.iterrows():
    if df.loc[index][0] != 'Estimate':
        state_name=df.loc[index][0]
    else:
        df.loc[index]['state']=state_name
        
# remove spacers
df.drop(df[df['Label'] !='Estimate'].index, inplace = True)
df.drop(labels=['Label'],axis=1, inplace = True)
df.to_csv('ACS_social_mod.csv', index=False)
df.head()
```

### Economic

```python
# Read in data and veiw
import pandas as pd
df=pd.read_csv('ACS_Economics.csv')
# Reformat rows
# Remove all percentage rows
df.drop(df[df['Label'] =='Percent'].index, inplace = True)



# Add state column
ls=[]
for index, row in df.iterrows():
    if df.loc[index][0] != 'Estimate':
        ls.append(df.loc[index][0])
    
    
# remove spacers
df.drop(df[df['Label'] !='Estimate'].index, inplace = True)
# create new coloumn
df['state']=ls
df.drop(labels=['Label'],axis=1, inplace = True)
df.to_csv('ACS_economics_mod.csv', index=False)
df.head()
```

# Queries

### Primary Dataset

1. Compare two states COVID deaths in Winter on same update time.

(join,where, order by, limit)

```python
%%bigquery
SELECT a.last_update, a.state, b.state, a.deaths, b.deaths from covid_19_staging.Winter_Cases as a full outer join covid_19_staging.Winter_Cases as b on a.last_update = b.last_update and a.state <> b.state where a.country = 'US' and a.deaths is NOT NULL and a.state is NOT NULL order by a.deaths desc limit 10;
```

2. Compare two states COVID deaths in Spring on same update time.

(join, where, order by, limit)

```python
%%bigquery
SELECT distinct a.last_update, a.state, b.state, a.deaths, b.deaths from covid_19_staging.Spring_Cases as a join covid_19_staging.Spring_Cases as b on a.last_update = b.last_update and a.state <> b.state where a.country = 'US' and a.deaths is NOT NULL and a.state is NOT NULL order by a.deaths desc limit 10;
```

3. Compare Texas and California confirmed cases and deaths over time.

((join, where, order by, limit)

```python
%%bigquery
SELECT a.last_update, a.deaths as Texas_Deaths, b.deaths as California_Deaths, a.confirmed as Texas_Confirmed,b.confirmed as California_Confirmed from covid_19_staging.Winter_Cases as a join covid_19_staging.Winter_Cases as b on a.last_update = b.last_update and a.state <> b.state where a.country = 'US' and a.state='Texas' and b.state='California' and a.deaths is NOT NULL and a.state is NOT NULL order by a.last_update desc limit 10;
```

4. Find and sum Covid Deaths over 10 in select states. Show state + sum of deaths, order by alpha

(where, group by,having, order by)

```python
%%bigquery
select state, sum(deaths) as total_deaths from covid_19_staging.Winter_Cases where state='Texas' or state='Oregon' or state='Arizona' or state='New York' group by state having sum(deaths)>0 order by state;
```

5. Sum number of total cases per state. Display states above 1,000 cases ordered by total number of cases.

(where, group by,having, order by)

```python
%%bigquery
select state, sum(confirmed) as total_cases_winter from covid_19_staging.Winter_Cases where country='US' group by state having sum(confirmed)>1000 order by total_cases_winter desc limit 10;
```

## Secondary Dataset

1. SELECT top 10 states where unemployment rate is gretaer t
(where, order by, limit, join)

```python
%%bigquery
SELECT demo.state,econ.Median_family_income, demo.White,demo.Black_or_African_American,demo.American_Indian_and_Alaska_Native,demo.Asian,demo.Hispanic_or_Latino
FROM acs_staging.demo_and_housing as demo INNER JOIN acs_staging.economics as econ on demo.state=econ.state
INNER JOIN acs_staging.housing_characteristics as house on demo.state = house.state
INNER JOIN acs_staging.social_characteristics as social on demo.state = social.state
WHERE econ.Unemployment_Rate>5.0
ORDER by econ.Median_family_income desc
limit 5
;
```

2. Select top 10 states where the %born in the U.S is less than %80. Displaying number of ag,construction workers and median earnings for that state.

(join, where, group by, order by, limit)

```python
%%bigquery
SELECT demo.state,econ.Agriculture_forestry,econ.Construction,econ.Median_earnings_for_workers,(social.Born_in_US/demo.Total_population) as percentage_born_in_US
FROM acs_staging.demo_and_housing as demo INNER JOIN acs_staging.economics as econ on demo.state=econ.state
INNER JOIN acs_staging.housing_characteristics as house on demo.state = house.state
INNER JOIN acs_staging.social_characteristics as social on demo.state = social.state
WHERE (social.Born_in_US/demo.Total_population) < 0.8
ORDER BY econ.Agriculture_forestry desc
LIMIT 5
```

3. Selecting top 10 states(ordered bymedian family income), and their median room size,avg family size and median family income WHEERE the avergae family size is greater than 2
(join, where, group by, order by, limit)

```python
%%bigquery
SELECT demo.state,house.Median_rooms, social.Average_family_size, econ.Median_family_income
FROM acs_staging.demo_and_housing as demo INNER JOIN acs_staging.economics as econ on demo.state=econ.state
INNER JOIN acs_staging.housing_characteristics as house on demo.state = house.state
INNER JOIN acs_staging.social_characteristics as social on demo.state = social.state
WHERE social.Average_family_size > 2
ORDER BY econ.Median_family_income
LIMIT 10
```

4. Select states, and thier percent without health insuance, and racial breakdowns WHERE the mean time to get to work is greater than 25 min
(Join,where, group by, having, order by, limit)

```python
%%bigquery
SELECT demo.state,(econ.No_health_insurance_coverage/econ.Population_over_16) as percent_without_health_insurance, econ.Mean_travel_time_to_work, demo.White,demo.Black_or_African_American,demo.American_Indian_and_Alaska_Native,demo.Asian,demo.Hispanic_or_Latino
FROM acs_staging.demo_and_housing as demo INNER JOIN acs_staging.economics as econ on demo.state=econ.state
INNER JOIN acs_staging.housing_characteristics as house on demo.state = house.state
INNER JOIN acs_staging.social_characteristics as social on demo.state = social.state
WHERE econ.Mean_travel_time_to_work > 25
ORDER BY (econ.No_health_insurance_coverage/econ.Population_over_16) desc
LIMIT 5
```

5. Selecting states, adn their avg mean family income WHERE the number of grandparents living at home with children is greater than 100,000

(Join, where, group by, order by)

```python
%%bigquery
SELECT demo.state, social.Grandparents as Grandparents_living_at_home, econ.Mean_family_income
FROM acs_staging.demo_and_housing as demo INNER JOIN acs_staging.economics as econ on demo.state=econ.state
INNER JOIN acs_staging.housing_characteristics as house on demo.state = house.state
INNER JOIN acs_staging.social_characteristics as social on demo.state = social.state
WHERE social.Grandparents>100000
ORDER BY social.Grandparents desc
LIMIT 5
```
