# Queries

### Primary Dataset

1. Compare two states COVID deaths in Winter on same update time.

(join,where, order by, limit)

```python
%%bigquery
SELECT a.last_update, a.state, b.state, a.deaths, b.deaths from covid_19_staging.Winter_Cases as a full outer join covid_19_staging.Winter_Cases as b on a.last_update = b.last_update and a.state <> b.state where a.country = 'US' and a.deaths is NOT NULL and a.state is NOT NULL order by a.deaths desc limit 10;
```

2. Compare two states COVID deaths in Spring on same update time.

(join, where, order by, limit)

```python
%%bigquery
SELECT distinct a.last_update, a.state, b.state, a.deaths, b.deaths from covid_19_staging.Spring_Cases as a join covid_19_staging.Spring_Cases as b on a.last_update = b.last_update and a.state <> b.state where a.country = 'US' and a.deaths is NOT NULL and a.state is NOT NULL order by a.deaths desc limit 10;
```

3. Compare Texas and California confirmed cases and deaths over time.

((join, where, order by, limit)

```python
%%bigquery
SELECT a.last_update, a.deaths as Texas_Deaths, b.deaths as California_Deaths, a.confirmed as Texas_Confirmed,b.confirmed as California_Confirmed from covid_19_staging.Winter_Cases as a join covid_19_staging.Winter_Cases as b on a.last_update = b.last_update and a.state <> b.state where a.country = 'US' and a.state='Texas' and b.state='California' and a.deaths is NOT NULL and a.state is NOT NULL order by a.last_update desc limit 10;
```

4. Find and sum Covid Deaths over 10 in select states. Show state + sum of deaths, order by alpha

(where, group by,having, order by)

```python
%%bigquery
select state, sum(deaths) as total_deaths from covid_19_staging.Winter_Cases where state='Texas' or state='Oregon' or state='Arizona' or state='New York' group by state having sum(deaths)>0 order by state;
```

5. Sum number of total cases per state. Display states above 1,000 cases ordered by total number of cases.

(where, group by,having, order by)

```python
%%bigquery
select state, sum(confirmed) as total_cases_winter from covid_19_staging.Winter_Cases where country='US' group by state having sum(confirmed)>1000 order by total_cases_winter desc limit 10;
```

## Secondary Dataset

1. Show top Deaths for California where income level is above 75,000 at distinct update times.
(where, order by, limit, join)

```python
%%bigquery
select distinct b.last_update, a.GEONAME, a.ESTIMATE, b.deaths from acs_staging.Income as a inner join covid_19_staging.Winter_Cases as b on a.GEONAME = "California" and b.state = "California" where a.ESTIMATE > 75000 and b.deaths is NOT NULL order by b.deaths desc limit 10;
```

2. Sums covid-19 deaths in each state over time. Displays state, income level above 25,000, and totalnumber of deaths

(join, where, group by, order by, limit)

```python
%%bigquery
select a.GEONAME, a.ESTIMATE, sum(b.deaths) as total_deaths from acs_staging.Income as a inner join covid_19_staging.Winter_Cases as b on a.GEONAME = b.state where a.ESTIMATE > 25000 and b.deaths is NOT NULL group by a.GEONAME, a.ESTIMATE order by sum(b.deaths) desc limit 10;
```

3. Selecting "Middle class" states (income level between 100,000 and 35,000) and displaying total number of cases and income level of the state

(join, where, group by, order by, limit)

```python
%%bigquery
select a.GEONAME, a.ESTIMATE, sum(b.confirmed) as total_cases from acs_staging.Income as a inner join covid_19_staging.Winter_Cases as b on a.GEONAME = b.state where a.ESTIMATE > 35000 and a.ESTIMATE < 100000 and b.deaths is NOT NULL group by a.GEONAME, a.ESTIMATE order by a.ESTIMATE desc, sum(b.confirmed) desc limit 10;
```

4. Select states that have a total covid case count over 2000 (for the spring), display cases,deaths, income level for state, ordered by total cases
(Join,where, group by, having, order by, limit)

```python
%%bigquery
select a.GEONAME, a.ESTIMATE,sum(b.deaths) as total_deaths, sum(b.confirmed) as total_cases from acs_staging.Income as a inner join covid_19_staging.Winter_Cases as b on a.GEONAME = b.state where b.deaths is NOT NULL group by a.GEONAME, a.ESTIMATE having sum(b.confirmed)>2000 order by sum(b.confirmed) desc limit 10;
```

5. Selecting 3 states (Texas+ California+ Ore) and comparing covid cases, deaths and income level for winter time.

(Join, where, group by, order by)

```python
%%bigquery
select a.GEONAME, a.ESTIMATE,sum(b.deaths) as total_deaths, sum(b.confirmed) as total_cases from acs_staging.Income as a inner join covid_19_staging.Winter_Cases as b on a.GEONAME = b.state where a.GEONAME='Texas' or a.GEONAME='California' or a.GEONAME='Oregon' and b.deaths is NOT NULL group by a.GEONAME, a.ESTIMATE  order by sum(b.confirmed) desc;
```
